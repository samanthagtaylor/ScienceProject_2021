```python
#### import pandas as pd
from Bio import SeqIO
import csv
import itertools
import datetime
#from itertools import as zip

# Python program to convert a list 
# to string using join() function 
    
# Function to convert   
def listToString(s):  
    
    # initialize an empty string 
    str1 = " " 
    
    # return string   
    return (str1.join(s))    


#Load in NCBI Data
coviddata=pd.read_csv('Sequences_02-13-2021_form.csv') 
coviddf = pd.DataFrame(coviddata)
#print(coviddf)

#Load in base sequence
for record in SeqIO.parse("BaseSequences.fa", "fasta"):
#for record in SeqIO.parse("sequences_covid_spike.fasta", "fasta"):
  base=record.seq
  #print(len(base))

variantdf = []

#For all the sequences in the NCBI set, look at each sequence and compare to the base sequence for mutations
for allrecord in SeqIO.parse("Sequences_02-13-2021.fasta", "fasta"):
#for allrecord in SeqIO.parse("BaseSequences.fasta", "fasta"):

  if len(allrecord.seq) >= 1273:
    allvariants=allrecord.seq 
    # How to use find() 
    if (allvariants.find('X') == -1): 
       # print (allvariants)
        a=[i for i in range(len(base)) if base[i] != allvariants[i]]
        res = [x + 1 for x in a] 
        listToStr = ' '.join([str(elem) for elem in res])
        #print(listToStr)
        #print(res)   
  id=allrecord.id
  idstring=id.split(".")
  variantdf.append([idstring[0], listToStr])

df = pd.DataFrame(variantdf, columns=['Accession', 'mutations'])  

joineddata=df.merge(coviddf,on='Accession',how='left')
#print(joinddata)


## Data Analysis

usa = joineddata["USA"]
usa_df = pd.DataFrame(usa)
USA_State_List = usa_df.USA.unique()
#print (USA_State_List)
#print(joineddata["Collection_Date"])

#Filtering based on date
joineddata['Collection_Data'] = pd.to_datetime(joineddata['Collection_Date'], format='%Y-%m-%d')
#filtered_df = joineddata.loc[(joineddata["Collection_Date"] <= '2020-04-1')]
#filtered_df = joineddata.loc[(joineddata["Collection_Date"] >= '2020-04-2')&(joineddata["Collection_Date"] <= '2020-06-1')]
#filtered_df = joineddata.loc[(joineddata["Collection_Date"] >= '2020-06-2')&(joineddata["Collection_Date"] <= '2020-08-1')]
#filtered_df = joineddata.loc[(joineddata["Collection_Date"] >= '2020-08-2')&(joineddata["Collection_Date"] <= '2020-10-1')]
#filtered_df = joineddata.loc[(joineddata["Collection_Date"] >= '2020-10-2')&(joineddata["Collection_Date"] <= '2020-12-1')]
filtered_df = joineddata.loc[(joineddata["Collection_Date"] >= '2020-12-2')&(joineddata["Collection_Date"] <= '2021-02-1')]


#Counting variants 
for x in USA_State_List:
  usa = filtered_df[filtered_df["USA"] == str(x)]
  usa.to_csv("mutations_ak.csv") 
  variant_count=usa['mutations'].value_counts()
  variant_count_df = pd.DataFrame(variant_count)
  #print(variant_count_df)
  index = variant_count_df.index
  number_of_rows = len(index)
  print(x, number_of_rows)
```
